```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# İris veri setini yükle
iris = load_iris()

# Veri setini bir DataFrame'e dönüştür
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Veriye bir göz atalım
print(df.head())
```

```python
# K-means algoritmasını uygula (3 küme, çünkü 3 tür var)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df)

# Kümeleri veri setine ekleyelim
df['Küme'] = kmeans.labels_

# Kümeleri inceleyelim
print(df.head())
```

```python
# Kümeleme sonuçlarını görselleştir
plt.figure(figsize=(8,6))
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['Küme'], cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.75, marker='x')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('K-means Kümeleme Sonuçları - İris Veri Seti')
plt.show()
```
